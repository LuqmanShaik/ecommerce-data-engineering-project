# 04 - Analytics SQL

## Objective
This notebook contains business analytics queries on the Gold Layer tables.

Source Tables:
- gold_sales
- gold_customers
- gold_products
- gold_sellers
- gold_delivery


```sql
DESCRIBE workspace.default.gold_sales;
```

```sql
SELECT
    ROUND(SUM(payment_value)) AS total_revenue
FROM workspace.default.gold_sales;
```

```sql
SELECT
COUNT(DISTINCT order_id) AS total_orders
FROM workspace.default.gold_sales
```

```sql
SELECT 
COUNT(DISTINCT customer_id) AS total_customers
FROM workspace.default.gold_sales;
```

```sql
SELECT 
ROUND(AVG(payment_value)) AS avg_revenued_order_value
FROM workspace.default.gold_sales
```

```sql
SELECT payment_type,
ROUND(SUM(payment_value)) AS total_revenue
FROM workspace.default.gold_sales
GROUP BY payment_type
ORDER BY total_revenue DESC
```

```sql
SELECT product_id,
ROUND(SUM(payment_value)) AS total_revenue
FROM gold_sales
GROUP BY product_id
ORDER BY total_revenue DESC
LIMIT 10;
```

```sql
SELECT
    gp.product_category_name_english,
    ROUND(SUM(gs.price), 2) AS revenue
FROM workspace.default.gold_sales gs
JOIN workspace.default.gold_products gp
ON gs.product_id = gp.product_id
GROUP BY gp.product_category_name_english
ORDER BY revenue DESC;
```

```sql
SELECT
    product_id,
    COUNT(*) AS total_sales
FROM workspace.default.gold_sales
GROUP BY product_id
ORDER BY total_sales DESC
LIMIT 10;
```

```sql
SELECT
    ROUND(AVG(price), 2) AS average_price
FROM workspace.default.gold_sales;
```

```sql
SELECT
    ROUND(AVG(freight_value), 2) AS average_freight_cost
FROM workspace.default.gold_sales;
```

```sql
SELECT
    seller_id,
    ROUND(SUM(price), 2) AS revenue
FROM workspace.default.gold_sales
GROUP BY seller_id
ORDER BY revenue DESC
LIMIT 10;
```

```sql
SELECT
    seller_id,
    COUNT(DISTINCT order_id) AS total_orders
FROM workspace.default.gold_sales
GROUP BY seller_id
ORDER BY total_orders DESC;
```

```sql
SELECT
    seller_state,
    COUNT(*) AS total_sellers
FROM workspace.default.gold_sellers
GROUP BY seller_state
ORDER BY total_sellers DESC;
```

```sql
SELECT
    s.seller_state,
    ROUND(SUM(gs.price), 2) AS revenue
FROM workspace.default.gold_sales gs
JOIN workspace.default.gold_sellers s
ON gs.seller_id = s.seller_id
GROUP BY s.seller_state
ORDER BY revenue DESC;
```

```sql
SELECT
    customer_state,
    COUNT(*) AS total_customers
FROM workspace.default.gold_customers
GROUP BY customer_state
ORDER BY total_customers DESC;
```

```sql
SELECT
    customer_city,
    COUNT(*) AS total_customers
FROM workspace.default.gold_customers
GROUP BY customer_city
ORDER BY total_customers DESC
LIMIT 10;
```

```sql
SELECT
    gc.customer_state,
    ROUND(SUM(gs.payment_value), 2) AS revenue
FROM workspace.default.gold_sales gs
JOIN workspace.default.gold_customers gc
ON gs.customer_id = gc.customer_id
GROUP BY gc.customer_state
ORDER BY revenue DESC;
```

```sql
SELECT
    ROUND(AVG(delivery_days), 2) AS average_delivery_days
FROM workspace.default.gold_delivery;
```

```sql
SELECT
    delivery_status,
    COUNT(*) AS total_orders
FROM workspace.default.gold_delivery
GROUP BY delivery_status
ORDER BY total_orders DESC;
```

```sql
SELECT
    ROUND(AVG(estimated_delivery_days), 2) AS average_estimated_delivery_days
FROM workspace.default.gold_delivery;
```

```sql
SELECT
    ROUND(AVG(review_score), 2) AS average_review_score
FROM workspace.default.gold_sales;
```

```sql
SELECT
    review_score,
    COUNT(*) AS total_reviews
FROM workspace.default.gold_sales
GROUP BY review_score
ORDER BY review_score;
```

```sql
SELECT
    order_status,
    COUNT(*) AS total_orders
FROM workspace.default.gold_sales
GROUP BY order_status
ORDER BY total_orders DESC;
```

```sql
SELECT
    YEAR(order_purchase_timestamp) AS year,
    MONTH(order_purchase_timestamp) AS month,
    COUNT(DISTINCT order_id) AS total_orders
FROM workspace.default.gold_sales
GROUP BY
    YEAR(order_purchase_timestamp),
    MONTH(order_purchase_timestamp)
ORDER BY
    year,
    month;
```

```sql
SELECT
    YEAR(order_purchase_timestamp) AS year,
    MONTH(order_purchase_timestamp) AS month,
    ROUND(SUM(payment_value), 2) AS revenue
FROM workspace.default.gold_sales
GROUP BY
    YEAR(order_purchase_timestamp),
    MONTH(order_purchase_timestamp)
ORDER BY
    year,
    month;
```
