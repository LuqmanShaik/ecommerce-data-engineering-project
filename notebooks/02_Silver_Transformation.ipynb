```python
customers_df = spark.table("workspace.default.bronze_customers")
customers_df.printSchema()
display(customers_df)
```

```python
print("Total Rows: ",customers_df.count())
print("Total Columns:", len(customers_df.columns))
```

```python
customers_df.dtypes
```

```python
from pyspark.sql.functions import col
for column in customers_df.columns:
    null_count = customers_df.filter(col(column).isNull()).count()
    print(f"{column} : {null_count}")
```

```python
total_rows = customers_df.count()
distinct_rows = customers_df.distinct().count()
print("Total Rows: ", total_rows)
print("Distinct Rows: ", distinct_rows)
print("Duplicate Rows: ", total_rows - distinct_rows)
```

```python
duplicate_customer_ids = (
    customers_df
    .groupBy("customer_id")
    .count()
    .filter("count > 1")
)
display(duplicate_customer_ids)
```

```python
silver_customers = (
    customers_df
    .drop_duplicates(["customer_id"])
)
```

```python
silver_customers.write \
    .mode("overwrite") \
        .saveAsTable("workspace.default.silver_customers")
```

```python
silver_customers = spark.table("workspace.default.silver_customers")
display(silver_customers)
```

```python
print("Total Bronze Rows:" , customers_df.count())
print("Total Silver Rows:" , silver_customers.count())
```

```python
orders_df = spark.table("workspace.default.bronze_orders")
orders_df.printSchema()
display(orders_df)
```

```python
print("Total Rows: ", orders_df.count())
print("Total Columns: ", len(orders_df.columns))
```

```python
from pyspark.sql.functions import col 
for column in orders_df.columns:
    null_count = orders_df.filter(col(column).isNull()).count()
    print(f"{column} : {null_count}")
```

```python
duplicate_orders = (
    orders_df
    .groupBy("order_id")
    .count()
    .filter("count > 1")
)
display(duplicate_orders)
```

```python
from pyspark.sql.functions import to_timestamp

silver_orders = (
    orders_df
    .withColumn(
        "order_purchase_timestamp",
        to_timestamp("order_purchase_timestamp")
    )
    .withColumn(
        "order_approved_at",
        to_timestamp("order_approved_at")
    )
    .withColumn(
        "order_delivered_carrier_date",
        to_timestamp("order_delivered_carrier_date")
    )
    .withColumn(
        "order_delivered_customer_date",
        to_timestamp("order_delivered_customer_date")
    )
    .withColumn(
        "order_estimated_delivery_date",
        to_timestamp("order_estimated_delivery_date")
    )
)
```

```python
silver_orders.printSchema()
```

```python
silver_orders.write \
    .mode("overwrite") \
        .saveAsTable("workspace.default.silver_orders")
```

```python
display(spark.table("workspace.default.silver_orders"))
```
