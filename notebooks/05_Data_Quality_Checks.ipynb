# 05 - Data Quality Checks

## Objective

This notebook validates the Bronze, Silver, and Gold layers by checking:

- Row counts
- Null values
- Duplicate records
- Primary key constraints
- Referential integrity

```sql
SELECT COUNT(*) AS bronze_customers 
FROM workspace.default.bronze_customers
```

```sql
SELECT COUNT(*) AS silver_customers
FROM workspace.default.silver_customers;
```

```sql
SELECT COUNT(*) AS gold_customers
FROM workspace.default.gold_customers;
```

```sql
SELECT
    (SELECT COUNT(*) FROM workspace.default.bronze_customers) AS bronze_count,
    (SELECT COUNT(*) FROM workspace.default.silver_customers) AS silver_count,
    (SELECT COUNT(*) FROM workspace.default.gold_customers) AS gold_count;
```

```sql
SELECT COUNT(*) AS null_customer_id
FROM workspace.default.silver_customers
WHERE customer_id IS NULL;
```

```sql
SELECT COUNT(*) AS null_product_id
FROM workspace.default.silver_products
WHERE product_id IS NULL;
```

```sql
SELECT COUNT(*) AS null_seller_id
FROM workspace.default.silver_sellers
WHERE seller_id IS NULL;
```

```sql
SELECT COUNT(*) AS null_order_id
FROM workspace.default.silver_orders
WHERE order_id IS NULL;
```

```sql
SELECT
    customer_id,
    COUNT(*) AS duplicate_count
FROM workspace.default.silver_customers
GROUP BY customer_id
HAVING COUNT(*) > 1;
```

```sql
SELECT
    order_id,
    COUNT(*) AS duplicate_count
FROM workspace.default.silver_orders
GROUP BY order_id
HAVING COUNT(*) > 1;
```

```sql
SELECT
    product_id,
    COUNT(*) AS duplicate_count
FROM workspace.default.silver_products
GROUP BY product_id
HAVING COUNT(*) > 1;
```

```sql
SELECT
    seller_id,
    COUNT(*) AS duplicate_count
FROM workspace.default.silver_sellers
GROUP BY seller_id
HAVING COUNT(*) > 1;
```

```sql
SELECT
    COUNT(*) AS invalid_customer_keys
FROM workspace.default.silver_customers
WHERE customer_id IS NULL;
```

```sql
SELECT
    COUNT(*) AS invalid_product_keys
FROM workspace.default.silver_products
WHERE product_id IS NULL;
```

```sql
SELECT
    COUNT(*) AS invalid_seller_keys
FROM workspace.default.silver_sellers
WHERE seller_id IS NULL;
```

```sql
SELECT
    COUNT(*) AS invalid_order_keys
FROM workspace.default.silver_orders
WHERE order_id IS NULL;
```

```sql
SELECT COUNT(*) AS missing_customers
FROM workspace.default.gold_sales gs
LEFT JOIN workspace.default.gold_customers gc
ON gs.customer_id = gc.customer_id
WHERE gc.customer_id IS NULL;
```

```sql
SELECT COUNT(*) AS missing_products
FROM workspace.default.gold_sales gs
LEFT JOIN workspace.default.gold_products gp
ON gs.product_id = gp.product_id
WHERE gp.product_id IS NULL;
```

```sql
SELECT COUNT(*) AS missing_sellers
FROM workspace.default.gold_sales gs
LEFT JOIN workspace.default.gold_sellers s
ON gs.seller_id = s.seller_id
WHERE s.seller_id IS NULL;
```

```sql
SELECT COUNT(*) AS negative_prices
FROM workspace.default.gold_sales
WHERE price < 0;
```

```sql
SELECT COUNT(*) AS negative_freight
FROM workspace.default.gold_sales
WHERE freight_value < 0;
```

```sql
SELECT COUNT(*) AS invalid_review_scores
FROM workspace.default.gold_sales
WHERE review_score NOT BETWEEN 1 AND 5;
```

```sql
SELECT COUNT(*) AS future_purchase_dates
FROM workspace.default.gold_sales
WHERE order_purchase_timestamp > CURRENT_TIMESTAMP();
```

```sql
SELECT COUNT(*) AS invalid_delivery_dates
FROM workspace.default.gold_delivery
WHERE order_delivered_customer_date < order_purchase_timestamp;
```

```sql
SELECT
    'Row Count Validation' AS validation_type,
    'Completed' AS status

UNION ALL

SELECT
    'Null Check',
    'Completed'

UNION ALL

SELECT
    'Duplicate Check',
    'Completed'

UNION ALL

SELECT
    'Primary Key Validation',
    'Completed'

UNION ALL

SELECT
    'Referential Integrity',
    'Completed'

UNION ALL

SELECT
    'Business Rule Validation',
    'Completed';
```
